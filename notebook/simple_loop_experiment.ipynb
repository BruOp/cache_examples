```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
output = pd.read_csv('../output/simple_loop.csv', header=None)
```

```python
steps = output[0]
num_iterations = output[1]
timings = output[2]
```

```python
x = range(1, len(steps) + 1)
plt.plot(x, (timings / num_iterations))
plt.scatter(x=x, y=(timings / num_iterations))
# plt.boxplot(timings)

# adding horizontal grid lines
plt.grid(True)
plt.xticks(x, steps)
plt.title('Time per Iteration vs Step Size')
plt.xlabel('Step Size (Bytes)')
plt.ylabel('Average Time per Iteration (ns)')
plt.rcParams['figure.figsize'] = [12, 8]
plt.show()
```
