```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
output = pd.read_csv('../output/simple_loop.csv', header=None)

```

```python
steps = output.iloc[0:,0]
num_iterations = output.iloc[0:,1]
timings = output.iloc[0:,2]
```

```python
num_iterations
```

```python
x = range(len(steps))
plt.plot((timings / num_iterations))
plt.scatter(x=x, y=(timings / num_iterations))


# adding horizontal grid lines
plt.grid(True)
plt.xticks(x, steps * 4)
plt.title('Time per Iteration vs Step Size')
plt.xlabel('Step Size (Bytes)')
plt.ylabel('Average Time per Iteration (ns)')
plt.rcParams['figure.figsize'] = [12, 8]
plt.show()
```
