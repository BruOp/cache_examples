```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
N_STEPS = 32 # The number of times the program iterates per sample
# Each row is (size, [sample_timing])
output = pd.read_csv('../output/simple_loop.csv', header=None)
sizes = output.iloc[0:,0]

```

```python
sizes
```

```python
timings = output.transpose().iloc[1:,:]
```

```python
for i in range(len(sizes)):
    timing = output.iloc[i, 1:] / N_STEPS
    size = sizes[i]
    
    plt.hist(timing, bins=50)
    plt.xlabel('Time (ns)')
    plt.ylabel('Frequency')
    plt.title(f'Timing with Step Size of {i}, {len(timing)} Samples')
    plt.grid(True)
    plt.show()
```

```python
# plot violin plot
plt.violinplot(output.iloc[:,1:] / N_STEPS,
                   showmeans=False,
                   showmedians=True)
# plt.boxplot(output.iloc[:,1:])

# adding horizontal grid lines
plt.grid(True)
plt.xticks(range(1,12), sizes * 8)
plt.title('Time per Iteration vs Step Size')
plt.xlabel('Step Size (Bytes)')
plt.ylabel('Average Time per Iteration (ns)')
plt.rcParams['figure.figsize'] = [12, 8]
```
