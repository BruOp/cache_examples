```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
output = pd.read_csv('../output/matrix.csv', header=None)
```

```python
sizes = output[0]
naive_timings = output[1] / 1e6
transpose_timings = output[2] / 1e6
```

```python
x = range(len(sizes) - 4)
timings = naive_timings[0:7]
plt.plot(timings)
plt.scatter(x=x, y=timings)

timings = transpose_timings[0:7]
plt.plot(timings)
plt.scatter(x=x, y=timings)


# adding horizontal grid lines
plt.grid(True)
plt.xticks(x, sizes)
plt.title('Matrix Multiplication')
plt.xlabel('Matrix Size')
plt.ylabel('Total Time (ns)')
plt.rcParams['figure.figsize'] = [12, 8]
plt.legend(('Naive Multiplication', 'Transpose First'))
plt.show()
```

```python
x = range(len(sizes))
timings = np.log10(naive_timings)
plt.plot(timings)
plt.scatter(x=x, y=timings)

timings = np.log10(transpose_timings)
plt.plot(timings)
plt.scatter(x=x, y=timings)


# adding horizontal grid lines
plt.grid(True)
plt.xticks(x, sizes)
plt.title('Matrix Multiplication')
plt.xlabel('Matrix Size')
plt.ylabel('Total Time (ns)')
plt.rcParams['figure.figsize'] = [12, 8]
plt.legend(('Naive Multiplication', 'Transpose First'))
plt.show()
```

```python
x = range(len(sizes))
timings = transpose_timings / naive_timings
plt.plot(timings)
plt.scatter(x=x, y=timings)


# adding horizontal grid lines
plt.grid(True)
plt.xticks(x, sizes)
plt.title('Speed of Transpose vs Naive Matrix Multiplication')
plt.xlabel('Matrix Size')
plt.ylabel('Factor Speed Up')
plt.rcParams['figure.figsize'] = [12, 8]
plt.show()
```
